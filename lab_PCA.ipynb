```python
df = pd.read_csv("data.csv")
```

```python
df.head()
```

```python
df_2 = df.loc[:, 'id':'fractal_dimension_worst']
```

```python
df_2.describe()
```

```python
df_2.iloc[:, 1].unique()
```

```python
df_2['diagnosis'].replace('B', 'Benign', inplace= True)
df_2['diagnosis'].replace('M', 'Malignant', inplace= True)
```

```python
df_2
```

```python
feature_df = df_2.iloc[:,2:]
label_df = df_2.iloc[:,1]
```

```python
label_df
```

```python
from sklearn.preprocessing import StandardScaler
x = feature_df.values
x = StandardScaler().fit_transform(x)
```

```python
import numpy as np 
np.mean(x), np.std(x)
```

```python
x[0]
```

```python
from sklearn.decomposition import PCA

pca_breast = PCA(n_components=2)
```

```python
principle_components = pca_breast.fit_transform(x)
```

```python
pbc_df = pd.DataFrame(data=principle_components, columns=['PC_1', 'PC_2'])
```

```python
pbc_df
```

```python
pca_breast.explained_variance_ratio_
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure
plt.figure(figsize=(10,10))
plt.xlabel('PC_1')
plt.ylabel('PC_2')

targets = ['Benign', 'Malignant']

colors = ['b','r']

for target, color in zip(targets, colors):
    indicesToKeep = label_df == target
    plt.scatter(pbc_df.loc[indicesToKeep,'PC_1'],pbc_df.loc[indicesToKeep,'PC_2'], c = color)
    
```

```python
data = pd.read_csv("gisette_train.data",header=None, delim_whitespace=True)
```

```python
label_df = pd.read_csv("gisette_train.labels",header=None, delim_whitespace=True)
```

```python
label_df.head()
```

```python
data.head()
```

```python
data.shape
```

```python
from sklearn.preprocessing import StandardScaler
x = data.values
x = StandardScaler().fit_transform(x)
```

```python
import numpy as np 
np.mean(x), np.std(x)
```

```python
from sklearn.decomposition import PCA

pca_gisette = PCA(n_components=2)
```

```python
principle_components = pca_gisette.fit_transform(x)
```

```python
pbc_df = pd.DataFrame(data=principle_components, columns=['PC_1', 'PC_2'])
```

```python
pbc_df
```

```python
pca_gisette.explained_variance_ratio_
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure
plt.figure(figsize=(10,10))
plt.xlabel('PC_1')
plt.ylabel('PC_2')

targets = [1, -1]

colors = ['b','r']

for target, color in zip(targets, colors):
    indicesToKeep = label_df[0] == target
    plt.scatter(pbc_df.loc[indicesToKeep,'PC_1'],pbc_df.loc[indicesToKeep,'PC_2'], c = color)
    
```
